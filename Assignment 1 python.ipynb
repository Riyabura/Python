```python
import numpy as np
from sklearn.cluster import KMeans

# Step 1: Convert the data set to (x, y) coordinate array
X = [1, 1, 1, 10, 10, 10]
Y = [2, 4, 0, 2, 4, 0]
data = np.array(list(zip(X, Y)))


# Step 2: Cluster the data set into two clusters using K-means
kmeans = KMeans(n_clusters=2, random_state=0).fit(data)
labels = kmeans.labels_

print(data)
print(labels)
```

**Example 3**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Convert the data set to (x, y) coordinate array
x = [4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]
data = np.array(list(zip(x, y)))
data

# Step 2: Plot the scatter diagram
plt.scatter(x, y, color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()

# Step 3: Use the elbow method to determine the number of clusters
inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

# Print the values of inertias
print(inertias)

# Plot the elbow curve (graph to decide the value of k)
plt.plot(range(1, 11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

# Use K-means clustering to create clusters
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

# Plot the clustered data
plt.scatter(x, y, c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, marker='*', c='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('K-means Clustering')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Import the CSV file into a DataFrame
df = pd.read_csv(r'C:\Users\hp\Downloads\Mall_Customers.csv')

# Check the head of the data
print(df.head())

# Create a scatter plot of Spending Score vs. Annual Income
plt.scatter(df['Spending Score (1-100)'], df['Annual Income (k$)'])
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Annual Income (k$)')
plt.show()

# Use Elbow method to determine optimal number of clusters (K)
inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df[['Spending Score (1-100)', 'Annual Income (k$)']])
    inertias.append(kmeans.inertia_)

plt.plot(range(1, 11), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()

# Use KMeans algorithm with optimal number of clusters
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(df[['Spending Score (1-100)', 'Annual Income (k$)']])

# Create a scatter plot with color-coded clusters
plt.scatter(df['Spending Score (1-100)'], df['Annual Income (k$)'], c=kmeans.labels_, cmap='rainbow')
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Annual Income (k$)')
plt.show()
```
